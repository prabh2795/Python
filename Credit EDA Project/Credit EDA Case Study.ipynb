```python
# Comparing EXT_SOURCE_2 and EXT_SOURCE_3 column 
round(df_current.loc[:, ['EXT_SOURCE_2', 'EXT_SOURCE_3']].describe(),3)
```

We can see that the difference between mean and median is quite narrow for both the columns. Also, the variation from 25th to 75th percentile is equally distributed. Thus, let's impute the missing values by the mean value of EXT_SOURCE_2 and EXT_SOURCE_3 respectively.

```python
# Imputing EXT_SOURCE_2 and EXT_SOURCE_3 by mean values
df_current.loc[np.isnan(df_current['EXT_SOURCE_2']), ['EXT_SOURCE_2']] = df_current['EXT_SOURCE_2'].mean()
df_current.loc[np.isnan(df_current['EXT_SOURCE_3']), ['EXT_SOURCE_3']] = df_current['EXT_SOURCE_3'].mean()
```

```python
#Finding out the % number of missing values in each column after imputation
(round(100*(df_current.isnull().sum()/len(df_current.index)),2))
```

We see the standard deviation of AMT_GOODS_PRICE is very high. Also there are good amount of outliers in the colum as well. So, imputing missing values with mean or median will be biased in this case. Hence, it is better to remove the rows with missing AMT_GOODS_PRICE. Infact these rows are only 0.09% of the whole records.

```python
df_current = df_current[~np.isnan(df_current['AMT_GOODS_PRICE'])]
```

We have 5 columns for the number of enquiries to Credit Bureau about the client. It doesn't make much sense to count the number for a very short period of time such as an hour or a day or a week before the application date. On the other had it also make lesser sense if we count the number for a very long period of time. Hence, it will be feasible for us if we take the count for a quater, which is not too short or too long time period. So, we have to do two things now:-

1. We will handle missing data for the column AMT_REQ_CREDIT_BUREAU_QRT

2. We will delete the column for the count of hour, day, week, month and year.

```python
# Handling AMT_REQ_CREDIT_BUREAU_QRT 
df_current['AMT_REQ_CREDIT_BUREAU_QRT'].describe()
```

This is an integer type variable, and thus have values 0, 1, 2 etc. We cannot impute the NaNs by the mean or the median (1.53 enquires does not make sense!). Thus, we need to impute them by the mode - the most common occurring value if that approach is feasible.

```python
# converting AMT_REQ_CREDIT_BUREAU_QRT to type 'category'
df_current['AMT_REQ_CREDIT_BUREAU_QRT'] = df_current['AMT_REQ_CREDIT_BUREAU_QRT'].astype('category')

# displaying frequencies of each category
df_current['AMT_REQ_CREDIT_BUREAU_QRT'].value_counts()
```

The most common value of AMT_REQ_CREDIT_BUREAU_QRT is 0, so let's impute the NaNs by that.

```python
# Imputing AMT_REQ_CREDIT_BUREAU_QRT with 0
df_current.loc[pd.isnull(df_current['AMT_REQ_CREDIT_BUREAU_QRT']), ['AMT_REQ_CREDIT_BUREAU_QRT']] = 0
```

```python
# Deleting AMT_REQ_CREDIT_BUREAU_HOUR, AMT_REQ_CREDIT_BUREAU_DAY, AMT_REQ_CREDIT_BUREAU_WEEK, AMT_REQ_CREDIT_BUREAU_MON, AMT_REQ_CREDIT_BUREAU_YEAR
df_current = df_current.drop(['AMT_REQ_CREDIT_BUREAU_HOUR','AMT_REQ_CREDIT_BUREAU_DAY','AMT_REQ_CREDIT_BUREAU_WEEK','AMT_REQ_CREDIT_BUREAU_MON', 'AMT_REQ_CREDIT_BUREAU_YEAR'],axis=1)
```

### Removing outliers

Finding outliers with numerical columns and then removing those records these can be done using the 1.5 IQR Rule.

```python
col_list_outlier =  ['AMT_INCOME_TOTAL','AMT_CREDIT','AMT_ANNUITY','AMT_GOODS_PRICE','AGE']
for col in col_list_outlier:
    q1=df_current[col].quantile(0.25)
    q3=df_current[col].quantile(0.75)
    iqr=q3-q1
    low_range=q1-1.5*iqr
    high_range=q3+1.5*iqr
    df_current=df_current.loc[(df_current[col]>low_range)&(df_current[col]<high_range)]
df_current.shape
```

```python
# Percentage of rows lost
round((100 - 100*(len(df_current.index)/307511)),2)
```

We have lost almost 10% of the rows in the outlier handling exercise. But this is still acceptable as removal of outliers ensures a fair analysis of data.

### Binning of AGE, AMT_INCOME_TOTAL, AMT_CREDIT and EXT_SOURCE_SCORE columns

**Creating three age groups - Young, Mid age and Senior citizen**

```python
df_current['AGE'].describe()
```

```python
# Define function for categorizing AGE_GROUP (Young, Mid age and Senior)
def age_group(x):
    if (x < 40):
        return 'Young'
    elif (x >= 40 and x < 60):
        return 'Mid Age'
    else:
        return 'Senior'
```

```python
#Creating new column AGE_GROUP
df_current['AGE_GROUP'] = df_current['AGE'].apply(age_group)
df_current.head()
```

**Creating three credit amount groups - High, medium and low**

```python
round(df_current['AMT_CREDIT'].describe(),2)
```

```python
# Define function for categorizing CREDIT_GROUP (High, Medium and Low)
def credit_group(x):
    if (x < 500000):
        return 'Low'
    elif (x >= 500000 and x < 750000):
        return 'Medium'
    else:
        return 'High'
```

```python
#Creating new column CREDIT_GROUP
df_current['CREDIT_GROUP'] = df_current['AMT_CREDIT'].apply(credit_group)
df_current.head()
```

**Creating three income groups - Hign, medium and Low**

```python
df_current['AMT_INCOME_TOTAL'].describe()
```

```python
# Define function for categorizing AMT_INCOME_TOTAL (Low, Medium and High)
def income_group(x):
    if (x < 100000):
        return 'Low'
    elif(x >= 100000 and x < 150000):
        return 'Medium'
    else:
        return 'High'
```

```python
# Creating column INCOME_GROUP with AMT_INCOME_TOTAL (Low, Medium and High)
df_current['INCOME_GROUP'] = df_current['AMT_INCOME_TOTAL'].apply(income_group)
df_current.head()
```

For the simplification of analysis of Normalized score from external data source, we are taking average of two different external data sources (EXT_SOURCE_2 and EXT_SOURCE_3) and creating a column called **EXT_SOURCE_SCORE**

```python
df_current['EXT_SOURCE_SCORE'] = round(((df_current['EXT_SOURCE_2'] + df_current['EXT_SOURCE_3'])/2),2)
df_current.head()
```

```python
# Dropping (EXT_SOURCE_2 and EXT_SOURCE_3 columns
df_current.drop(['EXT_SOURCE_2','EXT_SOURCE_3'],axis=1,inplace=True)
```

```python
df_current.head()
```

**Creating three External source score groups - High, Medium and Low**

```python
df_current['EXT_SOURCE_SCORE'].describe()
```

```python
# Define function for categorizing EXT_SOURCE_SCORE (Low, Medium and High)
def ext_source_group(x):
    if (x < 0.4):
        return 'Low'
    elif (x >= 0.4 and x < 0.6):
        return 'Medium'
    else:
        return 'High'
```

```python
# Creating column EXT_SCORE_CATEGORY with High, medium and Low
df_current['EXT_SCORE_CATEGORY'] = df_current['EXT_SOURCE_SCORE'].apply(ext_source_group)
df_current.head()
```
